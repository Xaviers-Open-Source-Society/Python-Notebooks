<br>•Shell tools:
People think of Python as a scripting language because it is used for coding operating-system-oriented scripts. The programs are launched from console command lines and perform tasks such as processing text files and launching other programs.

Python is used to direct other application components. During hardware testing, Python programs call the components that give low-level access to a device. <h2>Downside of Python</h2>
The only observed downside to Python is that it's execution speed is not as fast as fully-compiled and lower level languages like C and C++. For some tasks such low level languages as they are more directly mapped to the underlying architecture.Python has been optimized numerous times, and Python code runs fast enough by itself in most application domains. This depends if the speed of execution of the code matters. Furthermore, whenever you do something “real” in a Python script, like processing a file or constructing a graphical user interface (GUI), your program will actually run at C speed, since such tasks are immediately dispatched to compiled C code inside the Python interpreter.

<h2>How is Python Developed and Supported?</h2>
Python has a large and active development community that responds to issues and develops enhancements with efficient speed. Python developers work online with a source-control system. Changes are developed per a formal protocol, which includes writing a PEP (Python Enhancement Proposal) or other document, and extensions to Python’s regression testing system.
<h2>Open Source Tradeoffs</h2>
Having said that, it’s important to note that while Python enjoys a vigorous development community, this comes with inherent tradeoffs. Open source software can also appear chaotic and even resemble anarchy at times, and may not always be as smoothly implemented as the prior paragraphs might imply. Open Source exchange works according to the personal preference of the current set of developers which may or may not be the same as the user.
<h2>Strengths of Python</h2>
It's Object-Oriented and Functional:<br>
Python is entirely an object oriented language. It's classes support polymorphism,operator overloading and multiple inheritance.Besides serving as a powerful code structuring and reuse device, Python’s OOP nature makes it ideal as a scripting tool for other object-oriented systems languages. For example, with the appropriate glue code, Python programs can subclass (specialize) classes implemented in C++, Java, and C#. Python can use both procedural as well as Object-Oriented programming modes. In recent times Python can also perform functional programming.<br>

•It is Free:<br>
Python is available for free use and distribution on the Internet. There are no restrictions on copying it, embedding it in your systems, or shipping it with your products. In fact, you can even sell Python’s source code, if you are so inclined. Python software is supported as well. Python online community responds to queries as fast as any commercial software company would. Moreover, because Python comes with complete source code, it empowers developers, leading to the creation of a large team of implementation experts. 
<br>
•It is Portable: <br>
The standard implementation of Python is written in portable ANSI C, and it compiles and runs on virtually every major platform currently in use. For example, Python programs run today on everything from PDAs to supercomputers. As a partial list, Python is available on: 
<br>• Linux and Unix systems
<br>• Microsoft Windows (all modern flavors) 
<br>• Mac OS (both OS X and Classic) 
<br>• BeOS, OS/2, VMS, and QNX 
<br>• Real-time systems such as VxWorks
<br>• Cray supercomputers and IBM mainframes
<br>• PDAs running Palm OS, PocketPC, and Linux 
<br>• Cell phones running Symbian OS, and Windows Mobile
<br>• Gaming consoles and iPods
<br>• Tablets and smartphones running Google’s Android and Apple’s iOS 
<br>Like the language interpreter itself, the standard library modules that ship with Python are implemented to be as portable across platform boundaries as possible. Further, Python programs are automatically compiled to portable byte code, which runs the same on any platform with a compatible version of Python installed.
<br>
 •It is Powerful: <br>
 Its toolset places it between traditional scripting languages (such as Tcl, Scheme, and Perl) and systems development languages (such as C, C++, and Java). Python provides all the simplicity and ease of use of a scripting language, along with more advanced software-engineering tools typically found in compiled languages.

 Python allows components of another programming language to access its contents and accesses the compnents of other programming languages. Python can be used inside environments of other systems. this also makes Python good at rapid prototyping—systems may be implemented in Python first, to leverage its speed of development, and later moved to C for delivery, one piece at a time, according to performance demands.
